```python
import os
import cv2
import random
import shutil
import string
```

```python
def Converting_file_of_Videos_to_file_of_images(path , fileName = "Data"):
    try:
            if not (os.path.exists(fileName)):
                os.makedirs(fileName)
                
        
    except OSError:
        print("Error in creating a directory ( ",fileName," ) ")
                
            
    for NbVideo , video in enumerate(os.listdir(path)):
        cam = cv2.VideoCapture(os.path.join(path, video))

            
        current_frame = 0 

        while(True):
            
            rat, frame = cam.read()
            
            if rat:
                
                name = os.path.join(fileName, "Video_" +  str(NbVideo) + "_Frame_" + str(current_frame) + ".png"  ) #"./"+ "/frame" + str(NbVideo) + "_" + str(current_frame) + ".png"
                print(name + "   have been created") 
                cv2.imwrite(name, frame)  
                
                current_frame += 1
                
            else:
                break    
            
        cam.release()
    return fileName
        
    
    
def generate_random_string(length):
    letters = string.ascii_letters  # You can include digits or punctuation if needed
    return "A"+''.join(random.choice(letters) for _ in range(length))


def rearrange_files(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter out any hidden files (like .DS_Store on macOS)
    files = [f for f in files if not f.startswith('.')]
    
    # Shuffle the list of files (or you can sort them)
    random.shuffle(files)
    
    # Rename files with a new order
    for i, filename in enumerate(files):
        # Get file extension
        extension = filename.split(".")[-1]
        
        # Construct new filename (e.g., file_1.txt)
        new_filename = f"file_{i + 1}.{extension}"
        
        # Get the full path for original and new filenames
        src = os.path.join(folder_path, filename)
        dst = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(src, dst)
        
    print("Files have been rearranged.")


    


def split_the_videos(path_correct, path_incorrect, split_percentage = 0.85, FileName = "video_data"):
    
    correct_files = os.listdir(path_correct)
    incorrect_files = os.listdir(path_incorrect)
        
      
        
    print(correct_files[0:3], "correct")
    print(incorrect_files[0:3],"incorrect")    
        
    correct_files = random.sample(correct_files,len(correct_files))
    incorrect_files = random.sample(incorrect_files, len(incorrect_files))
    
    split_point_correct = int(len(correct_files) * split_percentage)    
    split_point_incorrect = int(len(incorrect_files) * split_percentage) 
    
    training_correct = correct_files[:split_point_correct]
    training_incorrect =  incorrect_files[:split_point_incorrect]
    testing_correct = correct_files[split_point_correct:]
    testing_incorrect =  incorrect_files[split_point_incorrect:]
        
    
    
    if not(os.path.exists(FileName)):
        os.makedirs(FileName)
    train_path_correct = os.path.join(FileName,"training_correct")
    train_path_incorrect = os.path.join(FileName,"training_incorrect")
    test_path_correct = os.path.join(FileName,"test_correct")
    test_path_incorrect = os.path.join(FileName,"test_incorrect")     
    os.makedirs(train_path_correct)
    os.makedirs(train_path_incorrect)
    os.makedirs(test_path_correct)
    os.makedirs(test_path_incorrect)
    
    
    
         
    
    for file_name in training_correct:
        source_path = os.path.join(path_correct, file_name) # correct_images is the path to the diroctory with the correct images
        destination_path = os.path.join(train_path_correct, file_name) # training_dataset_correct is where we want to place the training correct images
        shutil.copyfile(source_path , destination_path)

    for file_name in training_incorrect:
        source_path = os.path.join(path_incorrect, file_name) 
        destination_path = os.path.join(train_path_incorrect, file_name) # testing_dataset_correct where we want to place the testing images
        shutil.copyfile(source_path , destination_path)
        
    for file_name in testing_correct:
        source_path = os.path.join(path_correct, file_name)
        destination_path = os.path.join(test_path_correct, file_name)
        shutil.copyfile(source_path , destination_path)

    for file_name in testing_incorrect:
        source_path = os.path.join(path_incorrect, file_name)
        destination_path = os.path.join(test_path_incorrect, file_name)
        shutil.copyfile(source_path , destination_path)    
    
    return  train_path_correct, train_path_incorrect, test_path_correct, test_path_incorrect     
        
        
            
        

   
```

```python
def create_training_testing_dataset(correct_images , incorrect_images , directory_that_saves_the_dataset , split_percentage = 0.9):
    try:
        if not(os.path.exists(directory_that_saves_the_dataset)):
            os.makedirs(directory_that_saves_the_dataset)
        else:
            directory_that_saves_the_dataset = generate_random_string(8)
            os.makedirs(directory_that_saves_the_dataset)
            
        training_dataset = os.path.join(directory_that_saves_the_dataset, "training")   
        testing_dataset = os.path.join(directory_that_saves_the_dataset, "testing")     
        os.makedirs(training_dataset)  
        os.makedirs(testing_dataset)  
        
        training_dataset_correct = os.path.join(training_dataset, "correct")
        training_dataset_incorrect = os.path.join(training_dataset, "incorrect")
        
        os.makedirs(training_dataset_correct)  
        os.makedirs(training_dataset_incorrect) 
        
        testing_dataset_correct = os.path.join(testing_dataset, "correct")
        testing_dataset_incorrect = os.path.join(testing_dataset, "incorrect")
        
        os.makedirs(testing_dataset_correct)  
        os.makedirs(testing_dataset_incorrect) 
        
    except OSError:
        print("an error occured!!")     
        
        
    # all the lines of code you see above is for creating the files that we will store images in
    
    # all the lines below is used to move the images from source to the appropriate files   
    
    # يرجغ اسامي الملفات الي جوا هذا الملف  
    correct_images_names = os.listdir(correct_images)
    incorrect_images_names = os.listdir(incorrect_images)
    # يغير مواقع الاسماء بشكل عشوائي
    correct_images_names_shuffeled = random.sample(correct_images_names, len(correct_images_names))
    incorrect_images_names_shuffeled = random.sample(incorrect_images_names, len(incorrect_images_names))
    
    # NOTE: there should be two lists one for correct data and one for incorrect
    
        
    split_point_for_correct = int( split_percentage * len(correct_images_names_shuffeled))    
    split_point_for_incorrect = int( split_percentage * len(incorrect_images_names_shuffeled))
    # جزء لترينينج و جزء لتيست
    
    # و المفروض هذا الحركه نسويها على كل من الصور الصحيحه و الخاطئ بس الصدق ماكفاني الوقت بكمل بكره الصبح ان شاء الله
    correct_training_images = correct_images_names_shuffeled[: split_point_for_correct]
    correct_testing_images = correct_images_names_shuffeled[split_point_for_correct:]
    
    incorrect_training_images = incorrect_images_names_shuffeled[: split_point_for_incorrect]
    incorrect_testing_images = incorrect_images_names_shuffeled[split_point_for_incorrect:]
    
    
    for file_name in correct_training_images:
        source_path = os.path.join(correct_images, file_name) # correct_images is the path to the diroctory with the correct images
        destination_path = os.path.join(training_dataset_correct, file_name) # training_dataset_correct is where we want to place the training correct images
        shutil.copyfile(source_path , destination_path)

    for file_name in correct_testing_images:
        source_path = os.path.join(correct_images, file_name) 
        destination_path = os.path.join(testing_dataset_correct, file_name) # testing_dataset_correct where we want to place the testing images
        shutil.copyfile(source_path , destination_path)
        
    for file_name in incorrect_training_images:
        source_path = os.path.join(incorrect_images, file_name)
        destination_path = os.path.join(training_dataset_incorrect, file_name)
        shutil.copyfile(source_path , destination_path)

    for file_name in incorrect_testing_images:
        source_path = os.path.join(incorrect_images, file_name)
        destination_path = os.path.join(testing_dataset_incorrect, file_name)
        shutil.copyfile(source_path , destination_path)    
        



def create_dataset_with_videos_spliting(path_correct, path_incorrect, split_percentage = 0.85 , Dataset_Name = "Dataset_PushUps"):

    train_path_correct , train_path_incorrect , test_path_correct, test_path_incorrect = split_the_videos(path_correct, path_incorrect, split_percentage = 0.85, FileName = "video_file")

    if not(os.path.exists(Dataset_Name)):
        os.makedirs(Dataset_Name)
        
    else:
        Dataset_Name = generate_random_string(length = 8)
        
    train_path = os.path.join(Dataset_Name, "training_data")
    test_path = os.path.join(Dataset_Name, "test_data")
    os.makedirs(train_path)
    os.makedirs(test_path)

    training_correct = os.path.join(train_path,"correct")
    training_incorrect = os.path.join(train_path,"incorrect")
    os.makedirs(training_correct)
    os.makedirs(training_incorrect)
    testing_correct = os.path.join(test_path,"correct")
    testing_incorrect = os.path.join(test_path,"incorrect")
    os.makedirs(testing_correct)
    os.makedirs(testing_incorrect)
    
    
    
    Converting_file_of_Videos_to_file_of_images(train_path_correct,training_correct)

    Converting_file_of_Videos_to_file_of_images(train_path_incorrect  , training_incorrect)

    Converting_file_of_Videos_to_file_of_images(test_path_correct ,testing_correct)

    Converting_file_of_Videos_to_file_of_images(test_path_incorrect , testing_incorrect)
    
    
    rearrange_files(training_correct )
    rearrange_files(training_incorrect)
    rearrange_files(testing_correct)
    rearrange_files(testing_incorrect)
    
        
       
```

```python
# first we will convert both the correct sequences and incorrect sequences to images instead of video
def Create_dataset_with_out_video_spliting(path_correct_videos, path_incorrect_videos, Dataset_name, Spliting_percentige = 0.85):

    if not(os.path.exists("correct_images") or os.path.exists("incorrect_images")):
        correct_images_fileName = "correct_images"
        incorrect_images_fileName = "incorrect_images"
        os.makedirs(correct_images_fileName)
        os.makedirs(incorrect_images_fileName)
    else:
        correct_images_fileName = generate_random_string(8)
        incorrect_images_fileName = generate_random_string(9)
        os.makedirs(correct_images_fileName)
        os.makedirs(incorrect_images_fileName)
            

    Converting_file_of_Videos_to_file_of_images(path_correct_videos ,correct_images_fileName)

    Converting_file_of_Videos_to_file_of_images(path_incorrect_videos, incorrect_images_fileName)

    cv2.destroyAllWindows()

    create_training_testing_dataset(correct_images_fileName , incorrect_images_fileName , "Dataset" , split_percentage = Spliting_percentige )

    print("correct images = ", len(os.listdir(correct_images_fileName)))
    print("incorrect images = ", len(os.listdir(incorrect_images_fileName)))
```
